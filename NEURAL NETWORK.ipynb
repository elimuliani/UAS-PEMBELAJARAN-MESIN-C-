```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('waterQuality1.csv')
```

```python
# menggunakan Pandas untuk menghitung jumlah nilai null (NaN) pada setiap kolom.
# berguna untuk mengetahui seberapa banyak data yang hilang atau kosong dalam setiap kolom. Jika outputnya 0, itu berarti tidak ada nilai null dalam kolom tersebut. 
data.isnull().sum()
```

```python
# Gantilah nilai '#NUM!' dengan NaN
data.replace('#NUM!', pd.NA, inplace=True)
```

```python
# Konversi kolom ke tipe numerik
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
# Gantilah nilai NaN dengan rata-rata kolom
data.fillna(data.mean(), inplace=True)
```

## Modelling

```python
# Pisahkan fitur (X) dan label (y)

x = np.asarray(data.drop(['is_safe'], axis = 1), dtype=np.int64)
y = np.asarray(data['is_safe'], dtype=np.int64)
```

```python
# Bagi data menjadi data latih dan data uji

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Inisialisasi model Neural Network
# Disesuaikan dengan jumlah layer dan neuron yang diperlukan

from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
```

```python
# Latih model
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Prediksi menggunakan data latih dan data uji
y_pred = model.predict(x_train)
print(accuracy_score(y_train, y_pred))

y_pred_test = model.predict(x_test)
print(accuracy_score(y_test, y_pred_test))
```

## Visualisasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualisasi distribusi beberapa parameter kualitas air
plt.figure(figsize=(12, 8))
```

```python
# Histogram parameter kualitas air
plt.subplot(2, 2, 1)
sns.histplot(data['aluminium'], bins=20, kde=True, color='blue')
plt.title('Distribution of Aluminium')

plt.subplot(2, 2, 2)
sns.histplot(data['ammonia'], bins=20, kde=True, color='green')
plt.title('Distribution of Ammonia')
```

```python
# Scatter plot parameter kualitas air terhadap label
plt.subplot(2, 2, 3)
sns.scatterplot(x='aluminium', y='ammonia', hue='is_safe', data=data, palette='coolwarm')
plt.title('Scatter Plot: Aluminium vs Ammonia')
```

```python
# Boxplot parameter kualitas air terhadap label
plt.subplot(2, 2, 4)
sns.boxplot(x='is_safe', y='aluminium', data=data, palette='Set2')
plt.title('Boxplot: Aluminium by Safety')

plt.tight_layout()
plt.show()
```
